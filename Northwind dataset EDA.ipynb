```python
df_customers= pd.read_csv(customers_path)
df_orders= pd.read_csv(orders_path)
df_products = pd.read_csv(products_path)
df_orderdetails = pd.read_csv(order_details)
```

```python
product_sales = df_orderdetails.groupby("ProductID")["Quantity"].sum().reset_index().head(5)
product_sales
```

```python
#Order table merged
complete_order_merge= df_orders.merge(df_orderdetails , how="inner", on="OrderID")
complete_order_merge
```

```python
#Average value by region
avaerage_order_value_by_region = complete_order_merge.groupby('ShipCountry')['Quantity'].mean().reset_index().head(5)
avaerage_order_value_by_region
```

```python
#Top 10 Products with the highest Unit Prices
Products_with_the_highest_Unit_Prices= df_products.set_index("ProductName").sort_values("UnitPrice", ascending = False).head(10)
Products_with_the_highest_Unit_Prices
```

```python
#Top 10 Products with the lowest Unit Prices
Products_with_the_lowest_Unit_Prices= df_products.set_index("ProductName").sort_values("UnitPrice", ascending = True).head(10)
Products_with_the_lowest_Unit_Prices
```

```python
customers_order_merged= df_customers.merge(complete_order_merge, on="CustomerID", how="inner")
customers_order_merged
```

```python
#top 5 customers by order
top_5_customers_by_order = customers_order_merged.set_index("ContactName").sort_values("Quantity").head(5)
top_5_customers_by_order
```

```python
#barchart of top 5 customers by order
top_5_customers_by_order.plot(kind="bar",stacked= True)
```

```python
products_customers_order_merged=customers_order_merged.merge(df_products,on="ProductID",how="inner")
products_customers_order_merged
```

```python
#top 5 products by order
top_5_products_by_order = products_customers_order_merged.set_index("ProductName").sort_values("Quantity").head(5)
top_5_products_by_order
```

```python
#barchart of top 5 products by order
top_5_products_by_order.plot(kind="bar",stacked = True)
```
