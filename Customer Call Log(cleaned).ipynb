```python
my_df= pd.read_excel("C:\\Users\\victor\\Documents\\csv_files\\Customer Call List.xlsx")
my_df
```

```python
my_df=my_df.drop_duplicates()
my_df
```

```python
my_df=my_df.drop(columns="Not_Useful_Column")
my_df
```

```python
my_df["Last_Name"]=my_df["Last_Name"].str.strip("123._/")
my_df
```

```python
#my_df["Phone_Number"]=my_df["Phone_Number"].apply(lambda x:str(x))
my_df["Phone_Number"]=my_df["Phone_Number"].str.replace("-","")
my_df["Phone_Number"]=my_df["Phone_Number"].str.replace("|","")
my_df["Phone_Number"]=my_df["Phone_Number"].str.replace("/","")
my_df
```

```python
my_df["Phone_Number"]=my_df["Phone_Number"].apply(lambda x:x[0:3]+"-"+ x[3:6]+"-"+x[6:10])
my_df
```

```python
my_df["Phone_Number"]=my_df["Phone_Number"].str.replace("--","")
my_df["Phone_Number"]=my_df["Phone_Number"].str.replace("Na--","")
```

```python
my_df
```

```python
my_df["Phone_Number"]=my_df["Phone_Number"].str.replace("Na","")
my_df["Phone_Number"]=my_df["Phone_Number"].str.replace("nan","")
my_df
```

```python
for x in my_df.index:
    if my_df.loc[x,"Phone_Number"]=="":
        my_df.drop(x,inplace=True)
my_df
```

```python
my_df[["Street","State","ZIP"]]=my_df["Address"].str.split(",",n=2,expand=True)
my_df
```

```python
my_df["Paying Customer"]=my_df["Paying Customer"].str.replace("Yes","Y")
my_df["Paying Customer"]=my_df["Paying Customer"].str.replace("No","N")
my_df["Do_Not_Contact"]=my_df["Do_Not_Contact"].str.replace("Yes","Y")
my_df["Do_Not_Contact"]=my_df["Do_Not_Contact"].str.replace("No","N")
my_df["Do_Not_Contact"]=my_df["Do_Not_Contact"].str.replace("NaN","N")
my_df
```

```python
for x in my_df.index:
    if my_df.loc[x,"Do_Not_Contact"]=="Y":
        my_df.drop(x,inplace=True)
my_df
```

```python
my_df=my_df.fillna("")
my_df
```
