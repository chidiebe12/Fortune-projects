```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
d_frame = pd.read_csv(r"C:\Users\Laptop\Documents\china_area_population_density_data.csv")
d_frame
```

```python
d_frame[["Area_s(Km2)","Percentage(Area%)"]]= d_frame["Area (km2)"].str.split(n=1,expand=True)
d_frame[["Population_s","Percentage(Population%)"]]= d_frame["Population"].str.split(n=1,expand=True)

d_frame
```

```python
d_frame["Population_s"]=d_frame["Population_s"].apply(lambda x: str(x))
d_frame["Area_s(Km2)"]=d_frame["Area_s(Km2)"].apply(lambda x: str(x))
d_frame["Area_s(Km2)"]=d_frame["Area_s(Km2)"].apply(lambda x: str(x))
d_frame["Population_s"]=d_frame["Population_s"].apply(lambda x: str(x))
```

```python
d_frame["Area_s(Km2)"]=d_frame["Area_s(Km2)"].str.replace(",","")
d_frame["Population_s"]=d_frame["Population_s"].str.replace(",","")
d_frame
```

```python
list(d_frame)
```

```python
d_frame.drop(columns="Unnamed: 0")
d_frame.drop(columns="Unnamed: 0_level_0")
d_frame
```

```python
d_frame["Population_s"]=pd.to_numeric(d_frame["Population_s"], errors='coerce').fillna(0).astype(int)
d_frame["Area_s(Km2)"]=pd.to_numeric(d_frame["Area_s(Km2)"], errors='coerce').fillna(0).astype(int)
```
