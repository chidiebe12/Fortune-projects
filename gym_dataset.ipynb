```python
import pandas as pd
import matplotlib.pyplot as plt 
import numpy as np
import os,shutil 
```

```python
df= pd.read_csv(r"C:\Users\Laptop\Documents\gym_members_exercise_tracking.csv")
```

```python
df
```

```python
df = df.rename(columns={"Session_Duration (hours)": "Session_Duration"})
```

```python
#dropping dupliacte values
df.drop_duplicates()
```

```python
#detecting missing values 
print(df.isnull().sum())
```

```python
#Avearge Calories burned by Experience level
avearge_Calories_burned_by_experience_level= df.set_index("Experience_Level").groupby("Experience_Level")["Calories_Burned"].mean().to_frame()
avearge_Calories_burned_by_experience_level                                                                                                            
```

```python
#Average Calories burned by session duration
average_Calories_burned_by_session_duration= df.groupby("Session_Duration")["Calories_Burned"].mean().to_frame().reset_index()
average_Calories_burned_by_session_duration
```

```python
plt.scatter(average_Calories_burned_by_session_duration['Session_Duration'],
            average_Calories_burned_by_session_duration['Calories_Burned'])
plt.xlabel('Session_Duration')
plt.ylabel('Calories_Burned')
plt.title('Relationship between Calories Burned and Session Duration')
plt.show()
```

```python
#filtering members who only burned more than 500 Calories
df_filtered = df[df["Calories_Burned"]>500]
df_filtered
```

```python
#Adding a new column("Average_Calories_Burned)
df["Average_Calories_Burned"]= df['Calories_Burned'].mean()
#sorting dataset by average calories burned 
dataset_sorted_by_Average_Calories_Burned = df.sort_values(by="Average_Calories_Burned", ascending=False)
dataset_sorted_by_Average_Calories_Burned
```

```python
df['Age_Group'] = pd.cut(df['Age'], bins=[18, 25, 35, 45, 55, float('inf')], 
                           labels=['18-24', '25-34', '35-44', '45-54','55 and above'], 
                           include_lowest=True)
```

```python
#average resting BPM for each age group.
average_resting_BPM_for_each_age_group = df.groupby("Age_Group")["Resting_BPM"].mean().reset_index()
average_resting_BPM_for_each_age_group
```

```python
plt.bar(average_resting_BPM_for_each_age_group["Resting_BPM"],average_resting_BPM_for_each_age_group["Age_Group"])
plt.xlabel("Resting_BPM")
plt.ylabel("Age_Group")
plt.title("average_resting_BPM_for_each_age_group")
plt.show()

plt.figure(figsize=(8,6))
```

```python
#Average Calories burned by age group
average_Calories_Burned_by_Age_group = df.groupby("Age_Group")["Average_Calories_Burned"].mean().reset_index()
average_Calories_Burned_by_Age_group
```

```python
#pivot table showing average calories burned by workout type 
pivot_table = df.pivot_table(index="Workout_Type", values = "Average_Calories_Burned")
pivot_table
```

```python
#data transformation from days/week to hours/week (2.5 hours per week)
df["Workout_Frequency (hours/week)"]= df["Workout_Frequency (days/week)"]*2.5
df
```
